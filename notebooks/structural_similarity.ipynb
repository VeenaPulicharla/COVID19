# Similarity caluclation method

```python
def caluclate_silimilarity(x,fp1):
    try:
        # print("input", x)
        mol2 = Chem.MolFromSmiles(x)
        fp2 = Chem.RDKFingerprint(mol2)
    except :
        return np.nan
    return DataStructs.TanimotoSimilarity(fp1,fp2)
```

# Similar structure drugs to Remdesivir

```python
anti_virals[anti_virals["Name"].isin(["Remdesivir"])]
```

```python
remdesivir_smiles = anti_virals[anti_virals["Name"].isin(["Remdesivir"])]["SMILES"]
remdesivir_smiles
```

```python
mol1 = Chem.MolFromSmiles(list(remdesivir_smiles)[0])
fp1 = Chem.RDKFingerprint(mol1)
```

```python
drug_df["Remdesivir_score"] = drug_df["SMILES"].apply(lambda x: caluclate_silimilarity(x, fp1))
```

```python
drug_df["Remdesivir_score"].isnull().sum()
```

```python
drug_df[drug_df.isnull().any(axis=1)]
```

```python
# similar structure drugs to Remdesivir
drug_df[drug_df["Remdesivir_score"]> 0.8][["Name","Remdesivir_score"]]
```

```python
# similarity between Vinflunine and Vinblastine
DataStructs.TanimotoSimilarity(fp1,fp3)
```

```python
# similarity between Vinflunine and Vinblastine with DiceSimilarity

DataStructs.FingerprintSimilarity(fp1,fp3, metric=DataStructs.DiceSimilarity)
```

```python
# similarity between Vinflunine and Vincristine

DataStructs.DiceSimilarity(fp1,fp2)
```
